a) DNS Root: `Servidor cuya funcion es traducir los nombres de dominio en las direcciones ip.`

b) DNS Top-Level-Domain: `Un TLD (top level domain), es el ultimo segmento de un dominio, por ejemplo (.com).`

c) DNS Authoritative: `El DNS autoritativo es el encargado de transformar el dominio en direcciones ip y enviarlo como respuesta ante la solicitud de servidores DNS recurrentes.` 

d) DNS Recursive: `Es la primera parada en una consulta de DNS, es un intermediario entre todos los DNS, primero envia una solicitud al DNS ROOT despues al TLD y luego uno al autoritativo para luego enviar una respuesta al cliente.`
