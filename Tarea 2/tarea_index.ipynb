### 1.- ¿Qué es cada uno de los siguientes tipos de DNS: servidor DNS raíz (Root), DNS de nivel superior (Top−Level−Domain), DNS autoritario (Authoritative), y servidor DNS recursivo (Recursive). Evidenciar una muestra o ejemplo de interrogación DNS para cada uno de estos servidores.

* DNS Root : Servidor cuya funcion es traducir los nombres de dominio en las direcciones ip .

* DNS Top-Level-Domain : Un TLD (top level domain) , es el ultimo segmento de un dominio , por ejemplo (.com).

* DNS Authoritative : El DNS autoritativo es el encargado de transformar el dominio en direcciones ip y enviarlo como respuesta ante la solicitud de servidores DNS recurrentes . 

* DNS Recursive : Es la primera parada en una consulta de DNS , es un intermediario entre todos los DNS , primero envia una solicitud al DNS ROOT despues al TLD y luego uno al autoritativo para luego enviar una respuesta al cliente .
